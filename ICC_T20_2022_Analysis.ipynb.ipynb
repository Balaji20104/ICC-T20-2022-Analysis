```python
#3. Most Winning Teams

team_win_count = WC['Winner'].value_counts()
plt.figure(figsize=(12, 5))
sns.barplot(x=team_win_count.index, y=team_win_count.values)
plt.title("Number of Matches Won by Each Team")
plt.xlabel("Team")
plt.ylabel("Number of Matches Won")
plt.xticks(rotation=90)
plt.show()
```

```python
# 4. Toss Winner vs Match Winner

toss_match_winner = WC[WC['Toss_winner'] == WC['Winner']].shape[0]
total_matches = WC.shape[0]

print(f"Toss winner also won the match in {toss_match_winner} out of {total_matches} matches.")
print(f"That’s {(toss_match_winner / total_matches) * 100:.2f}% of the matches.")
```

```python
# 5. Toss Decision Impact on Match Outcome

plt.figure(figsize=(8, 6))
sns.countplot(x='Toss_winner', hue='Winner', data=WC)
plt.title("Relationship between Toss Winner and Match Winner")
plt.xlabel("Team")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.legend(title="Match Winner", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
